## Logistic Regression Workshop

```python
import pandas as pd 
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score 

df = pd.read_csv('iris-data-clean.csv')
print(df.to_string())
```

```python
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
print(df.shape)
```

```python
#Replace the values 
df['class'] = df['class'].replace({'Setosa':0, 'Virginica':1, 'Versicolor':2})
print("Update Class values:")
print(df['class'].unique())
```

```python
print(df['class'].value_counts())
```

```python
df.head()
```

```python
df.tail()
```

```python
#Feature (x) and Target (y)
logReg = LogisticRegression()
x = df[['sepal_length_cm', 'sepal_width_cm', 'petal_length_cm', 'petal_width_cm']]
y = df['class']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(x_train.head())
```

```python
print(y_test)
```

```python
print(x_train.shape)
```

```python
print(x_test)
```

```python
logReg.fit(x_train, y_train)
```

```python
#Validate the model 
y_pred = logReg.predict(x_test)
print('Y predicted values:', y_pred)
```

```python
print(y_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(f"Confusion Matrix:\n{cm}")
```

```python
logReg.predict(x_test)
```

```python
#Prediction 
logReg.predict([[5.1, 3.5, 1.4, 0.2]])  
```

```python
import matplotlib.pyplot as plt

# Filter by class
df_0 = df[df['class'] == 0]  # Setosa
df_1 = df[df['class'] == 1]  # Virginica
df_2 = df[df['class'] == 2]  # Versicolor

# Scatter plot
fig, ax = plt.subplots(figsize=(10, 5))
ax.scatter(df_0['sepal_length_cm'], df_0['sepal_width_cm'], c='orange', label='Setosa (0)')
ax.scatter(df_1['sepal_length_cm'], df_1['sepal_width_cm'], marker='s', c='blue', label='Virginica (1)')
ax.scatter(df_2['sepal_length_cm'], df_2['sepal_width_cm'], marker='^', c='green', label='Versicolor (2)')

ax.set_xlabel('Sepal Length (cm)')
ax.set_ylabel('Sepal Width (cm)')
ax.set_title('Iris Dataset Scatter Plot by Class (Sepal Features)')
ax.legend()
plt.show()
```
